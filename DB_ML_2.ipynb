```python
import patsy
```

```python
y,X = patsy.dmatrices('Default ~ BILL_AMT1+BILL_AMT2+BILL_AMT3+BILL_AMT4+BILL_AMT5+BILL_AMT6+PAY_AMT1+PAY_AMT2+PAY_AMT3+PAY_AMT4+PAY_AMT5+PAY_AMT6+LIMIT_BAL',data=default)

```

```python
y = np.ravel(y)
```

```python
X
```

```python
from sklearn.linear_model import LogisticRegression as logistic
```

```python
yhat = logistic().fit(X,y).predict(X)
```

```python
pd.Series(yhat).value_counts()
```

```python
default['Default'].value_counts()
```

```python
from sklearn.neighbors import KNeighborsClassifier as knn
```

```python
pd.Series(knn().fit(X,y).predict(X)).value_counts()
```

```python
logistic().fit(X,y).score(X,y)
```

```python
knn().fit(X,y).score(X,y)
```

```python
default['yhat_knn'] = knn().fit(X,y).predict(X)
```

```python
default
```

```python
default.pivot_table(index='Default',columns='yhat_knn',values='ID',aggfunc='count')/len(default)
```

```python
for i in range(1,5):
    print(i,knn(n_neighbors=i).fit(X,y).score(X,y))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
for i in range(24,48,4):
    print(i,cross_val_score(knn(n_neighbors=i),X,y,cv=5).mean())
```

```python
from sklearn.ensemble import RandomForestClassifier as rf
```

```python
cross_val_score(rf(),X,y,cv=5).mean()
```

```python
for i in range(2,6):
    print(i,cross_val_score(rf(max_depth=1),X,y,cv=5).mean())
```

```python
import pandas as pd             # data package
import matplotlib.pyplot as plt # graphics 
import datetime as dt           # date tools, used to note current date  

# these are new 
import os                       # operating system tools (check files)
import requests, io             # internet and input tools  
import zipfile as zf            # zip file tools 
import shutil                   # file management tools
```

```python
url = 'http://files.grouplens.org/datasets/movielens/ml-latest-small.zip'
r = requests.get(url) 
```

```python
mlz = zf.ZipFile(io.BytesIO(r.content)) 
```

```python
movies = pd.read_csv(mlz.open('ml-latest-small/movies.csv'))
```

```python
ratings = pd.read_csv(mlz.open('ml-latest-small/ratings.csv'))
```

```python
movies
```

```python
ratings
```

```python
movies['genres'].str.split('|').str[0].unique()
```

```python
for i in movies['genres'].str.split('|').str[0].unique():
    movies[i] = 0
    movies.loc[movies['genres'].str.contains(i),i] = 1
```

```python
movies
```

```python
from sklearn.cluster import KMeans as kmeans
```

```python
movies['k30'] = kmeans(n_clusters=30).fit_predict(movies.loc[:,'Adventure':'War'])
```

```python
movies
```

```python
movies.loc[movies['k30']==3,'title']
```

```python
ratings2 = ratings.merge(movies,on='movieId',how='inner')
```

```python
ratings2
```

```python
top_movies = ratings2['movieId'].value_counts().iloc[:600].index
```

```python
top_users = ratings2['userId'].value_counts().iloc[:200].index
```

```python
ratings3 = ratings2.loc[(ratings2['movieId'].isin(top_movies))&(ratings2['userId'].isin(top_users)),:]
```

```python
ratings3
```

```python
piv = ratings3.pivot_table(index='title',columns='userId',values='rating',aggfunc='count').fillna(0)
```

```python
piv
```

```python
piv['k100'] = kmeans(n_clusters=100).fit_predict(piv.loc[:,1:610])
```

```python
piv.loc[piv['k100']==5,:].index
```

```python
movies.set_index('title').loc[:,'Adventure':'War']
```

```python
piv2 = piv.drop('k100',1).merge(movies.set_index('title').loc[:,'Adventure':'War'],
                         left_index=True,right_index=True,how='inner')
```
