```python
def  pocket_change(p,n,d,q):
    dollar_quarter = int(q/4)
    remain_quarter = ((q/4)%1) * 100
    dollar_dime = int(d/10)
    remain_dime = ((d/10)%1) * 100
    dollar_nickel = int(n/20)
    remain_nickel = ((n/20)%1) * 100
    dollar_pennies = int(p/100)
    remain_pennies = ((p/100)%1) * 100
    
    total_dollar = dollar_quarter+ dollar_dime + dollar_nickel + dollar_pennies
    total_cents = remain_quarter + remain_dime + remain_nickel + remain_pennies
    
    total_cents_to_dollar = int(total_cents/100) 
    remain_total_cents = ((total_cents/100)%1) 
    
    total_dollar = total_dollar + total_cents_to_dollar
    
    dollar = total_dollar + remain_total_cents
    return float(dollar)
      
```

```python
print('$ ', pocket_change(1,2,3,4))
```

##### Create  pocket_change function  and  first find the dollar value of quarter - integer part - using int(q/4)  and then extract the remaining quarter - decimal part - using ((q/4)%1) * 100 , Do this for dime , nickel and pennies by dividing by 10, 20 and 100 instead of 4 .  Now find the dollar value by summing up all the integer part of dollars for all  quarter, dime , nickel and pennies.  Find the cents remaining by summing up the decimal part of quarter, dime , nickel and pennies.   
#####  Extract the dollar value - integer part- in the cents remaining  using  int(total_cents/100) and also find the cents remaining  - decimal part - using ((total_cents/100)%1)  . 
#####  Now the total dollar (total_dollar) value will be the sum of initial dollar value and the dollar value from the cents remaining . Now a string (dollar ) is created to concatenate the dollar value and cents . This  string is then converted to float and function returns this float .  print () is used to print  dollar sign and the float value

```python
def remove_paren(paren_string):
    index1 = str(paren_string).find('(')
    index2 = str(paren_string).find(')')
    return paren_string[0:index1] + paren_string[index2+1 :]    
```

```python
paren = ['When we went to town on Monday (it was so early!) everyone was falling asleep', 'It is obvious (so obvious) that I am going to get a good grade on this exam','As she sang her famous line (the one they wrote that article about) the crowd went wild'] 
```

```python
for i in paren:
    print(remove_paren(i))
```
