```python
def angle(H,M):
    total_mins = H*60 + M
    hour_angle = total_mins*.5
    minute_angle = M*6
    angle= hour_angle-minute_angle
    if angle<0:
        angle = 360 + angle
    if angle>180:
        angle = 360 - angle
    if angle==360:
        angle = 0
    return angle
```
