```python
y = ratings2.loc[ratings2['title']=='Toy Story 2 (1999)','rating'].value_counts(normalize=True)
y = pd.DataFrame(y)
y
```

```python
x.rename(columns={'rating':'TS_1'}).merge(y.rename(columns={'rating':'TS_2'}),
        left_index=True,
        right_index=True,
        how='outer').fillna(0).plot()
```

```python
m_list = movies.loc[movies['title'].str.startswith('Back to the Future'),'title'].unique()
```

```python
x = pd.DataFrame(ratings2.loc[ratings2['title']==m_list[0],
                              'rating'].value_counts(normalize=True)).rename(columns={'rating':m_list[0]})
for i in range(1,len(m_list)):
    x = x.merge(pd.DataFrame(ratings2.loc[ratings2['title']==m_list[i],
                              'rating'].value_counts(normalize=True)).rename(columns={'rating':m_list[i]}),
            left_index=True,right_index=True,how='outer').fillna(0)
x.plot(figsize=(6,6))
```

```python
toy_stories = movies.loc[movies['title'].str.contains('Toy Stor'),'title'].unique()
for i in toy_stories:
    ratings2.loc[ratings2['title']==i,'rating'].value_counts(normalize=True).sort_index().plot.bar()
```
