```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
%matplotlib inline
```

```python
election = pd.read_csv(
    'https://raw.githubusercontent.com/mwaugh0328/Data_Bootcamp_Fall_2017/master/data_bootcamp_1127/trump_data.csv',
    encoding='latin-1')
```

```python
election
```

```python
reg = smf.ols('trump_share ~ income',data=election).fit()
```

```python
print(reg.summary())
```

```python
reg.params['income']*10000
```

```python
reg = smf.ols('trump_share ~ np.log(population)',data=election).fit()
print(reg.summary())
```

```python
np.log(election['population']).plot.hist(bins=30)
```

```python
election['population'].sort_values()
```

```python
reg = smf.ols('trump_share ~ np.log(population)+income',data=election).fit()
print(reg.summary())
```

```python
reg.params['income']*10000
```

```python
election[['population','income']].corr()
```

```python
election.columns
```

```python
reg = smf.ols('trump_share ~ np.log(population)+income+StateName',data=election).fit()
print(reg.summary())
```

```python
reg.params['income']*10000
```

```python
state_level = election.groupby('StateName').agg({'income':np.mean,'population':np.sum,'trump_share':np.mean})
```

```python
state_level
```

```python
reg = smf.ols('trump_share ~ income + np.log(population)',data=state_level).fit()
print(reg.summary())
```

```python
reg.params['income']*10000
```
