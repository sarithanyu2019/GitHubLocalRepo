#### Problem 3

```python
jobs['year'] = pd.DatetimeIndex(jobs['post_date']).year
jobs
```

```python
annual_stats = jobs.groupby(['year','state']).mean()
annual_stats
```

##### Qn a

```python
annual_stats.set_index('salary' , append=True , inplace = True)
annual_stats
```

##### Qn b

```python
annual_stats = annual_stats.loc[annual_stats.index.get_level_values('state').isin(['NY' , 'CA' , 'TX' ] ) , :]
annual_stats = annual_stats.reset_index('salary')
annual_stats['salary'] = np.round(annual_stats['salary'], decimals=2 )
annual_stats.set_index('salary' , append=True , inplace = True)
annual_stats
```

##### Qn c

```python
annual_stats = annual_stats.reset_index('salary')
annual_stats = annual_stats.reset_index('year')
annual_stats
```

```python
annual_stats.pivot_table(index='state',columns='year',values='salary').rename_axis(None, axis=0).to_excel('pivot_salary.xlsx')
```

##### Qn d

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig, ax = plt.subplots()
annual_stats_plot = annual_stats.loc[annual_stats['year'] == 2016]
annual_stats_plot.reset_index().sort_values(by='salary').plot.barh(x='state' , y = 'salary' , ax=ax,figsize=(8,3), width = 0.84, color='c' ) 
ax.set_title('Salaries for each state in 2016', loc='center' , fontsize=20)
ax.set_xlabel('Salary' , fontsize=15)
```

```python
ecommerce.rename(columns={ 'Avg. Session Length' : 'sess_len',
                           'Time on App' : 'app_time' ,
                           'Time on Website' : 'web_time' , 
                           'Length of Membership' : 'membership_len' ,
                           'Yearly Amount Spent' : 'yr_amtspent' }, inplace=True )
ecommerce
```

##### the Histograms for all variables are  normally distributed . So here no need to consider log model.

```python
reg = smf.ols('yr_amtspent ~ sess_len+app_time+web_time+membership_len',data=ecommerce).fit()
print(reg.summary())
```
